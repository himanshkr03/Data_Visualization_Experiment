```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Read the dataset from file into a DataFrame
file_path = '/content/phone_usage_india.csv'
data = pd.read_csv(file_path)

# Display the first few rows to verify the dataset is loaded
print("Dataset Preview:")
print(data.head())
```

```python
# Step 1: Create a bar chart using matplotlib
def create_bar_chart(df):
    plt.figure(figsize=(10, 6))
    df.groupby('Phone Brand')['User ID'].count().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Number of Users by Phone Brand', fontsize=16)
    plt.xlabel('Phone Brand', fontsize=12)
    plt.ylabel('Number of Users', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Additional bar chart for another column
    plt.figure(figsize=(10, 6))
    df.groupby('Primary Use')['User ID'].count().plot(kind='bar', color='lightgreen', edgecolor='black')
    plt.title('Primary Use Distribution', fontsize=16)
    plt.xlabel('Primary Use', fontsize=12)
    plt.ylabel('Number of Users', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

create_bar_chart(data)
```

```python
# Step 2: Generate a histogram using seaborn
def create_histogram(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Screen Time (hrs/day)'], kde=True, color='purple', bins=20)
    plt.title('Distribution of Screen Time', fontsize=16)
    plt.xlabel('Screen Time (hrs/day)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Additional histogram for another column
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Monthly Recharge Cost (INR)'], kde=True, color='orange', bins=20)
    plt.title('Distribution of Monthly Recharge Cost', fontsize=16)
    plt.xlabel('Monthly Recharge Cost (INR)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

create_histogram(data)

```

```python
# Step 3: Create a box plot using matplotlib
def create_box_plot(df):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Gender', y='Screen Time (hrs/day)', data=df, palette='Set2')
    plt.title('Screen Time by Gender', fontsize=16)
    plt.xlabel('Gender', fontsize=12)
    plt.ylabel('Screen Time (hrs/day)', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Additional box plot for another column
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Primary Use', y='Data Usage (GB/month)', data=df, palette='Set3')
    plt.title('Data Usage by Primary Use', fontsize=16)
    plt.xlabel('Primary Use', fontsize=12)
    plt.ylabel('Data Usage (GB/month)', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.show()

create_box_plot(data)

```

```python
def create_pie_chart(df):
    plt.figure(figsize=(8, 8))
    df['OS'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue'], startangle=140, explode=(0.1, 0), shadow=True)
    plt.title('OS Distribution', fontsize=16)
    plt.ylabel('')  # Hides y-axis label
    plt.show()

    # Additional pie chart for another column
    plt.figure(figsize=(8, 8))
    # Get the number of unique values in the 'Gender' column
    num_gender_categories = len(df['Gender'].value_counts())

    # Create an explode tuple with the appropriate length
    explode = (0.1,) + (0,) * (num_gender_categories - 1)

    df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'lightgreen'], startangle=140, explode=explode, shadow=True)
    plt.title('Gender Distribution', fontsize=16)
    plt.ylabel('')  # Hides y-axis label
    plt.show()


create_pie_chart(data)
```

```python
# Analysis and Summary
def analyze_effectiveness():
    print("\nAnalysis and Summary of Visualizations:")
    print("1. Bar Chart: Clear representation of categorical data like Phone Brand and Primary Use. Easy to identify the most popular categories.")
    print("2. Histogram: Effective for visualizing the distribution of continuous data such as Screen Time and Monthly Recharge Cost. Provides insights into the spread and skewness of data.")
    print("3. Box Plot: Highlights data distribution and outliers. Gender-based screen time and primary use-based data usage comparisons are insightful.")
    print("4. Pie Chart: Useful for showing proportions, such as OS and Gender distribution, but may not be ideal for detailed comparisons.")

analyze_effectiveness()
```
