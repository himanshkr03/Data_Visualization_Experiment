# **Line Chart**

```python
def line_chart(data, x_col, y_col, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data[x_col], data[y_col], marker='o', linestyle='-', color='b', label=f"{y_col} over {x_col}")
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend()
    plt.grid(True)
    plt.show()
def scatter_plot(data, x_col, y_col, hue_col, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=x_col, y=y_col, hue=hue_col, palette='viridis', s=100)
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend()
    plt.show()
def scatter_plot(data, x_col, y_col, hue_col, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=x_col, y=y_col, hue=hue_col, palette='viridis', s=100)
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend()
    plt.show()
def heatmap(data, title):
    # Select only numeric columns for correlation calculation
    numeric_data = data.select_dtypes(include=np.number)
    correlation_matrix = numeric_data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title(title)
    plt.show()
def plot_3d(data, x_col, y_col, z_col, color_col, title):
    fig = px.scatter_3d(data, x=x_col, y=y_col, z=z_col, color=color_col, title=title, size=color_col, opacity=0.8)
    fig.update_layout(scene=dict(
        xaxis_title=x_col,
        yaxis_title=y_col,
        zaxis_title=z_col
    ))
    fig.show()

```
